```python
import os
import sys

sys.path.append('/home/ubuntu/diffmoog_base/src')

import torch
import torchaudio
from synth.synth_constants import synth_constants
import glob

signal_transform = torchaudio.transforms.MelSpectrogram(sample_rate=synth_constants.sample_rate,
                                                        n_fft=400, hop_length=160, n_mels=80,
                                                        power=1.0, center=True)
```

```python
files = os.listdir('/home/ubuntu/diffmoog_base/data/modular/train/wav_files')

mel_max = -1e9
mel_min = 1e9

for i, f in enumerate(files):
    audio = torchaudio.load('/home/ubuntu/diffmoog_base/data/modular/train/wav_files/' + f)[0]
    mel = signal_transform(audio).abs()
    log_mel = torch.log(mel + 1e-9)
    
    mel_min = min(mel_min, log_mel.min())
    mel_max = max(mel_max, log_mel.max())
    
    if i % 2000 == 0:
        print(f'{i}/{len(files)}, min: {mel_min:.4f}, max: {mel_max:.4f}')
```
